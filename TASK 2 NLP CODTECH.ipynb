```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
import gensim.downloader as api
```

```python
text_data="""
We are a professional Business and Management Consulting firm committed to helping businesses achieve their goals.
 Our team of experts has years of experience in the industry and is dedicated to providing tailored solutions to each client.
  Explore our website to learn more about our services and how we can help your business grow"""
```

```python
#tokenization
tokens=word_tokenize(text_data.lower())
print("tokenization:")
print(tokens)
```

```python
#stop word removal
stop_words=set(stopwords.words('english'))
filtered_tokens=[word for word in tokens if word not in stop_words]
print("\nAfter Stop-Word removal:")
print(filtered_tokens)
```

```python
#stemming
porter_stemmer=PorterStemmer()
stemmed_tokens=[porter_stemmer.stem(word) for word in tokens]
print("\nAfter Stemming:")
print(stemmed_tokens)
```

```python
#Lemmatization
lemmatizer=WordNetLemmatizer()
lemmatized_tokens=[lemmatizer.lemmatize(word) for word in tokens]
print("\nAfter lemmatization:")
print(lemmatized_tokens)
```

```python
import gensim.downloader as api
word2vec_model = api.load("word2vec-google-news-300")
filtered_tokens = ["example", "words", "here"]
print("\nWord embeddings using Word2Vec:")
for word in filtered_tokens:
    if word in word2vec_model.key_to_index:
        print(f"{word}: {word2vec_model[word]}")
    else:
        print(f"{word} is not in the Word2Vec model's vocabulary.")
```

```python
#word embeddings using glove
glove_model=api.load("glove-wiki-gigaword-100")
print("\nWord embeddings using glove:")
for word in filtered_tokens:
  if word in glove_model.key_to_index:
    print(word,":",glove_model[word])
```
